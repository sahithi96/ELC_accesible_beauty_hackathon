```python
target_columns=['Tender Light',
'Desert Eden',
'Infinite Sky',
'Sensuous Stars',
'Paradise Moon',
'Radiant Mirage',
'Dream Dusk',
'Blushing Sands',
'Beautiful Magnolia',
'Modern Muse',
'Beautiful Belle',
'Beautiful',
'Estée Lauder Pleasures',
'Amber Mystique',
'Cinnabar',
'SpellBound']
x=data.drop(columns=target_columns)
y=data[target_columns]

train = data.iloc[:45000]
test = data.iloc[45000:]



xtrain=x.iloc[:45000]
ytrain=y.iloc[:45000]

xtest=x.iloc[45000:]
ytest=y.iloc[45000:]
```

```python

import numpy as np
import pandas as pd
from xgboost.sklearn import XGBClassifier
from sklearn.metrics import confusion_matrix


clf = XGBClassifier(max_depth=7, n_estimators=1000,
                   # objective='multi:softmax',num_class=13
                   )

model=clf.fit(xtrain, ytrain)
```

```python
import pickle
filename = 'global.pkl'
pickle.dump(model, open(filename, 'wb'))

# # to load the saved model
# bst = pickle.load(open(filename, 'rb'))
```

```python
n = 3
probas =  clf.predict_proba(xtest)
top_n_lables_idx = np.argsort(-probas)[:, :n]
top_n_probs = np.round(-np.sort(-probas),3)[:, :n]
colum_list=model.get_booster().feature_names
print(top_n_lables_idx)
print(top_n_probs)
top_n_labels = [colum_list[i] for i in top_n_lables_idx]

results = list(zip(top_n_labels, top_n_probs))

pd.DataFrame(results)
```
